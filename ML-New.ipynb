```python
sets = {}
class DataSet():
    # Rocket simulates a rocket ship for a game,
    #  or a physics simulation.
    
    def __init__(self):
        # Each rocket has an (x,y) position.
        self.features = None
        self.target = None
```

```python
def process_data(file):
    with open(file) as csvfile:
        readCSV = csv.reader(csvfile, delimiter=',')
        data = DataSet()
        data.features = []
        data.target = []
        for row in readCSV:
            data.features.append(list(np.float_(row[0:5])))
            data.target.append(int(row[5]))
    return data
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn import svm
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier

def run_model(data):
    X_train, X_test, y_train, y_test = train_test_split(data.features, data.target, test_size=0.4, random_state=0)
    clf_svm = svm.SVC(kernel='linear', C=1).fit(X_train, y_train)
    clf_knn = KNeighborsClassifier(n_neighbors=3)
    clf_nn = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1).fit(X_train, y_train)  
    y_pred_svm = clf_svm.fit(X_train, y_train).predict(X_test)
    y_pred_knn = clf_knn.fit(X_train, y_train).predict(X_test)
    y_pred_nn = clf_nn.fit(X_train, y_train).predict(X_test)
    print('svm', clf_svm.score(X_test, y_test))
    print('knn', clf_knn.score(X_test, y_test))
    print('nn', clf_nn.score(X_test, y_test))
    print("svm confusion matrix")
    print(confusion_matrix(y_test, y_pred_svm))
    print("knn confusion matrix")
    print(confusion_matrix(y_test, y_pred_knn))
    print("nn confusion matrix")
    print(confusion_matrix(y_test, y_pred_nn))
```
